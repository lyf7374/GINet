```python
import os
from utils.segment_and_save import segment_volume
import tqdm
# ===== 路径设置 =====
BASE_DIR = '/Users/yifanli/Desktop/P2'

# 模板和模板 mask
template_img_path = os.path.join(
    BASE_DIR, 'assets', 'templates', 'mni_icbm152_t1_tal_nlin_sym_09a.nii'
)
mask_template_img_path = os.path.join(
    BASE_DIR, 'assets', 'templates', 'mni_icbm152_t1_tal_nlin_sym_09a_mask.nii'
)

# 模型权重
model_weights = os.path.join(BASE_DIR, 'T0_Files', 'SG_v5_noc.pth')

# 输入和输出文件夹
IN_DIR = '/Users/yifanli/Desktop/P4/NeuroTestData/T1C'
OUT_DIR = '/Users/yifanli/Desktop/P4/NeuroTestData/GINetSeg'

os.makedirs(OUT_DIR, exist_ok=True)

# ===== 批量处理 =====
for fname in tqdm.tqdm(os.listdir(IN_DIR)):
    # 只处理 NIfTI 文件
    if not (fname.endswith('.nii') or fname.endswith('.nii.gz')):
        continue

    input_path = os.path.join(IN_DIR, fname)
    output_path = os.path.join(OUT_DIR, fname)  # 输出用同名文件，放在新目录下

    print(f'Processing: {input_path} -> {output_path}')

    try:
        segment_volume(
            input_path=input_path,
            output_path=output_path,
            model_weights='/Users/yifanli/Desktop/fromLinux/SG_v5t5_noc.pth',
            template_img_path=template_img_path,
            mask_template_img_path=mask_template_img_path,
            shapes=(192, 192, 192),
            norm_method='zs',
            n_patch=64,
            t_patch=64,
            device=None,  # None = 自动选择 CPU/GPU
        )
    except Exception as e:
        print(f'  [ERROR] Failed on {input_path}: {e}')
```
